```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import seaborn as sns
import matplotlib.pyplot as plt
import regex as re 
import glob
from collections import Counter
```

```python
df = pd.read_csv("Lyrik/Corpus_Explorer-master/explorer_stable/meta/Willatzen.tsv", sep="\t")
df.head(1)
```

```python
path = "Lyrik/Korpus_Lyrik/Willatzen/*.txt"
prefix = "Lyrik/Korpus_Lyrik/Willatzen/"

freq_list = []
for file in glob.glob(path):
    filename = file.replace(prefix, '').replace('.txt', '')
    with open(file, 'r') as f:
        data = f.read().replace('\n', '')
    
        
    freq_list.append(pd.Series(data, name=filename))
# print(freq_list)
```

```python
text = pd.DataFrame(freq_list)
# text
```

```python
text['ID'] = text.index
text.head(1)
```

```python
lyrik = text.merge(df, on='ID')
lyrik.drop('Unnamed: 0', axis=1, inplace=True)
lyrik.rename({0: 'text'}, axis=1, inplace=True)
# lyrik
```

```python
lyrik.to_csv('Lyrik/anthologien_csv/willatzen_anthologie_gedichte.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()

liste = []

for i, row in lyrik.iterrows():
    count_vect.fit_transform(row.text.split())
    titel = str(row.Autor)+'_'+str(row.Titel)
    dic = count_vect.vocabulary_
    frequencies = list(dic.values())
    words = list(dic.keys())
    liste.append(pd.Series(frequencies, words, name=titel))
```

```python
df = pd.DataFrame(liste)
df = df.fillna(0)
df = df.div(df.sum(axis=1), axis=0)
df
```

```python
path = "German_prosa/68_german_novels-master/*.txt"
prefix = "German_prosa/68_german_novels-master/"
    
text_list = []
for file in glob.glob(path):
    filename = file.replace(prefix, '').replace('.txt', '')
    with open(file, 'r') as f:
        data = f.read().replace('\n', ' ')
    
    text_list.append(pd.Series(data, name=filename))

prosa = pd.DataFrame(text_list)
```

```python
prosa['autor'] = '0'
prosa['titel'] = '0'

for i, row in prosa.iterrows():
    row['autor'] = i.split("_")[0]
    row['titel'] = i.split("_")[1]
prosa = prosa.rename(columns={0: "text"})
prosa.head()
```
