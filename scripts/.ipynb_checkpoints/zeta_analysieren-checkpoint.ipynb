```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
emotion100 = pd.read_csv('../results/zeta/emotion_100pro_wort.csv', header=None)
emotion100.columns = ['wort', 'zeta']
emotion100 = emotion100.sort_values(by='zeta', ascending=False)

emotion500 = pd.read_csv('../results/zeta/emotion_500pro_wort.csv', header=None)
emotion500.columns = ['wort', 'zeta']
emotion500 = emotion500.sort_values(by='zeta', ascending=False)

emotion1000 = pd.read_csv('../results/zeta/emotion_1000pro_wort.csv', header=None)
emotion1000.columns = ['wort', 'zeta']
emotion1000 = emotion1000.sort_values(by='zeta', ascending=False)

fig, axs = plt.subplots(1,3, figsize=(30,40))
sns.barplot(x='zeta', y=emotion100.wort[:10], data=emotion100, color='b', ax=axs[0])
axs[0].set_title('100 WÃ¶rter pro Segment')
sns.barplot(x='zeta', y=emotion500.wort[:10], data=emotion500, color='b', ax=axs[1])
axs[1].set_title('500 WÃ¶rter pro Segment')
sns.barplot(x='zeta', y=emotion1000.wort[:10], data=emotion1000, color='b', ax=axs[2])
axs[2].set_title('1000 WÃ¶rter pro Segment')
```
