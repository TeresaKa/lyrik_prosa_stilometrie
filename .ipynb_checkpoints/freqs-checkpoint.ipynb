```python
import glob
import regex as re
from collections import Counter

import pandas as pd
from scipy.stats import zscore
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def stop_word_removal(x):
    #token = x.split()
    german_stop_words = stopwords.words('german')
    for w in x:
        print('word',w)
    return #' '.join([w for w in x if w not in german_stop_words])
```

```python
poems = pd.read_csv('anthologien_csv/arent_anthologie_gedichte.csv')
german_stop_words = stopwords.words('german')
    
poems['text'] = [str(i).lower() for i in poems['text']] 
poems['removedstopword'] = poems['text'].apply(lambda x: ' '.join([item for item in str(x).split() if item not in german_stop_words]))
poems
```

```python
freq_list = []
count_vect = CountVectorizer()
for i, row in poems.iterrows():
    count_vect.fit_transform(row.removedstopword.split())
    titel = str(row.Autor) + '_' + str(row.Titel)
    vocab = count_vect.vocabulary_
    frequencies = list(vocab.values())
    words = list(vocab.keys())
    freq_list.append(pd.Series(frequencies, words, name=titel))
```

```python
counts = pd.DataFrame(freq_list)
counts = counts.fillna(0)
counts = counts.div(counts.sum(axis=1), axis=0)
counts.loc['Total_per_word'] = counts.sum()
counts = counts.sort_values(by='Total_per_word', axis=1, ascending=False)
counts.drop('Total_per_word', inplace=True, axis=0)

zscores = counts.apply(zscore)
zscores.drop(zscores.columns[1000:], inplace=True, axis=1)
```
